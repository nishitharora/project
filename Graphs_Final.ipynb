```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from all sheets into a dictionary of DataFrames
excel_file = '/content/Intrinsic Final.xlsx'  # Replace with the path to your Excel file
sheet_names = ['ludhiana', 'sangrur', 'gurdaspur', 'hoshiarpur', 'muktsar']  # Specify the names of your sheets

dfs = {}
for sheet_name in sheet_names:
    dfs[sheet_name] = pd.read_excel(excel_file, sheet_name=sheet_name)

# Combine all DataFrames into one DataFrame
combined_df = pd.concat(dfs.values(), ignore_index=True)

# Calculate average milk production per breed
average_production = combined_df.groupby('breed')['milkproductionperday'].mean()

# Calculate range of milk production per breed
max_production = combined_df.groupby('breed')['milkproductionperday'].max()
min_production = combined_df.groupby('breed')['milkproductionperday'].min()

# Calculate standard deviation of milk production per breed
std_production = combined_df.groupby('breed')['milkproductionperday'].std()

# Plotting
plt.figure(figsize=(10, 6))

# Bar plot for average production
bar = plt.bar(average_production.index, average_production.values, color='b', label='Average Production')

# Range plot for max and min production
plt.errorbar(average_production.index, average_production.values,
             yerr=[average_production.values - min_production.values, max_production.values - average_production.values],
             fmt='o', color='black', ecolor='gray', capsize=5, label='Production Range')

# Adding standard deviation as error bars
plt.errorbar(average_production.index, average_production.values, yerr=std_production.values,
             fmt='.', color='red', ecolor='orange', capsize=5, label='Standard Deviation')

# Annotate bars with average values
for i, val in enumerate(average_production.values):
    plt.text(i, val, f'{val:.2f}', ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Breed')
plt.ylabel('Milk Production (units)')
plt.title('Average, Range, and Standard Deviation of Milk Production by Breed')
plt.legend()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read data from all sheets into a dictionary of DataFrames
excel_file = '/content/Intrinsic Final.xlsx'  # Replace with the path to your Excel file
sheet_names = ['ludhiana', 'sangrur', 'gurdaspur', 'hoshiarpur', 'muktsar']  # Specify the names of your sheets

dfs = {}
for sheet_name in sheet_names:
    dfs[sheet_name] = pd.read_excel(excel_file, sheet_name=sheet_name)

# Combine all DataFrames into one DataFrame
combined_df = pd.concat(dfs.values(), ignore_index=True)

# Filter out rows where age is 0 or 19
combined_df = combined_df[(combined_df['age'] != 0) & (combined_df['age'] != 19)]

# Round age to nearest integer
combined_df['rounded_age'] = combined_df['age'].apply(lambda x: int(np.round(x)))

# Calculate average milk production per age
average_production = combined_df.groupby('rounded_age')['milkproductionperday'].mean()

# Calculate range of milk production per age
max_production = combined_df.groupby('rounded_age')['milkproductionperday'].max()
min_production = combined_df.groupby('rounded_age')['milkproductionperday'].min()

# Calculate standard deviation of milk production per age
std_production = combined_df.groupby('rounded_age')['milkproductionperday'].std()

# Plotting
plt.figure(figsize=(10, 6))

# Bar plot for average production
bar = plt.bar(average_production.index, average_production.values, color='b', label='Average Production')

# Range plot for max and min production
plt.errorbar(average_production.index, average_production.values,
             yerr=[average_production.values - min_production.values, max_production.values - average_production.values],
             fmt='o', color='black', ecolor='gray', capsize=5, label='Production Range')

# Adding standard deviation as error bars
plt.errorbar(average_production.index, average_production.values, yerr=std_production.values,
             fmt='.', color='red', ecolor='orange', capsize=5, label='Standard Deviation')

# Annotate bars with average values
for i, val in enumerate(average_production.values):
    plt.text(average_production.index[i], val, f'{val:.2f}', ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Milk Production (units)')
plt.title('Average, Range, and Standard Deviation of Milk Production by Age')
plt.legend()

# Set y-axis ticks to display as whole numbers for age
plt.xticks(range(int(min(average_production.index)), int(max(average_production.index)) + 1))

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from all sheets into a dictionary of DataFrames
excel_file = '/content/Intrinsic Final.xlsx'  # Replace with the path to your Excel file
sheet_names = ['ludhiana', 'sangrur', 'gurdaspur', 'hoshiarpur', 'muktsar']  # Specify the names of your sheets

dfs = {}
for sheet_name in sheet_names:
    dfs[sheet_name] = pd.read_excel(excel_file, sheet_name=sheet_name)

# Combine all DataFrames into one DataFrame
combined_df = pd.concat(dfs.values(), ignore_index=True)

# Define function to categorize farm sizes into sections
def categorize_farm_size(farm_size):
    if farm_size == 1:
        return '1'
    elif 2 <= farm_size <= 3:
        return '2-3'
    elif 4 <= farm_size <= 7:
        return '4-7'
    elif 8 <= farm_size <= 15:
        return '8-15'
    elif 16 <= farm_size <= 31:
        return '16-31'
    elif 32 <= farm_size <= 64:
        return '32-64'

# Create a new column for farm size sections
combined_df['farm_size_section'] = combined_df['farmsize'].apply(categorize_farm_size)

# Specify the order of the farm size sections
section_order = ['1', '2-3', '4-7', '8-15', '16-31', '32-64']

# Calculate average milk production per farm size section
average_production = combined_df.groupby('farm_size_section')['milkproductionperday'].mean()

# Calculate range of milk production per farm size section
max_production = combined_df.groupby('farm_size_section')['milkproductionperday'].max()
min_production = combined_df.groupby('farm_size_section')['milkproductionperday'].min()

# Calculate standard deviation of milk production per farm size section
std_production = combined_df.groupby('farm_size_section')['milkproductionperday'].std()

# Plotting
plt.figure(figsize=(10, 6))

# Bar plot for average production
bar = plt.bar(average_production.index, average_production.values, color='b', label='Average Production')

# Range plot for max and min production
plt.errorbar(average_production.index, average_production.values,
             yerr=[average_production.values - min_production.values, max_production.values - average_production.values],
             fmt='o', color='black', ecolor='gray', capsize=5, label='Production Range')

# Adding standard deviation as error bars
plt.errorbar(average_production.index, average_production.values, yerr=std_production.values,
             fmt='.', color='red', ecolor='orange', capsize=5, label='Standard Deviation')

# Annotate bars with average values
for i, val in enumerate(average_production.values):
    plt.text(average_production.index[i], val, f'{val:.2f}', ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Farm Size Section')
plt.ylabel('Milk Production (units)')
plt.title('Average, Range, and Standard Deviation of Milk Production by Farm Size Section')

# Set x-axis tick labels in specified order
plt.xticks(average_production.index, section_order)

plt.legend()

# Show plot
plt.show()
```
