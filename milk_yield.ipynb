```python
import pandas as pd

# Load the Excel file into a DataFrame
df = pd.read_excel('milkself10000.xlsx', sheet_name='Sheet1')

# Sort the DataFrame by 'fieldnum' and 'milk_yield' in descending order
df_sorted = df.sort_values(by=['farmnum', 'farmmilkpercattle'], ascending=[True, False])

# Initialize a dictionary to store the sums of top N highest yields for each field
top_n_sums = {}

# Specify the value of N (number of highest yields to sum)
N = 3  # You can change this to any desired value

# Loop through each field
for fieldnum, group in df_sorted.groupby('farmnum'):
    top_n_sums[fieldnum] = {}  # Initialize a dictionary to store sums for each field
    count = 0  # Initialize a counter to keep track of the top N yields
    # Iterate over the sorted group
    for index, row in group.iterrows():
        if count < N:
            if count + 1 not in top_n_sums[fieldnum]:
                top_n_sums[fieldnum][count + 1] = 0  # Initialize the sum for this position if not present
            top_n_sums[fieldnum][count + 1] += row['farmmilkpercattle']  # Add the yield to the sum for this position
            count += 1  # Increment the counter
        else:
            break  # Break the loop if we have already summed the top N yields

# Print the sum of the top N highest yields for each field individually
for fieldnum, sums in top_n_sums.items():
    print(f"Field {fieldnum}:")
    for position, total_sum in sums.items():
        print(f"Sum of top {position} highest yields:", total_sum)
    print()  # Add a newline for better readability
```
